###<a name='6'></a>*Logarithms*

#####In mathematics, a logarithm is the inverse operation to exponentiation.

#####The general form of a logarithmic function is:

$$
\log_b(x) = y \quad \text{if and only if} \quad b^y = x
$$

#####Where:
- *b* is the base of the logarithm,
- *x* is the argument,
- *y* is the result of the logarithm.

#### Important Properties of Logarithms:

1. **Logarithm with Base \( e \)** (Natural Logarithm):
   $$ \ln(x) \quad \text{is the logarithm to the base} \, e. $$

2. **Logarithm with Base 10**:
   $$ \log_{10}(x) \quad \text{is the logarithm to the base 10.} $$

3. **Logarithmic Identity for Base 1**:
   $$ \log_1(x) \quad \text{is undefined for any \( x \neq 1 \).} $$

4. **Zero Exponent Rule**:
   - $ x^0 = 1  $ for any $ x \neq 0 $.
   - The logarithm of 1 in any base is always 0:
     $$ \log_b(1) = 0 \quad \text{for any base} \, b. $$

5. **Inverse Property of Logarithms**:
   - Logarithms are the inverse of exponentiation:
     $$ \log_b(x) = y \quad \text{if and only if} \quad b^y = x. $$

6. **Product Rule**:
   $$ \log_b(xy) = \log_b(x) + \log_b(y) $$

7. **Quotient Rule**:
   $$ \log_b\left(\frac{x}{y}\right) = \log_b(x) - \log_b(y) $$

8. **Power Rule**:
   $$ \log_b(x^n) = n \cdot \log_b(x) $$

#### Example:
If $ 10^3 = 1000 $, then:
$$ \log_{10}(1000) = 3 $$

Similarly, if $ e^1 = e $, then:
$$ \ln(e) = 1 $$

###<a name='7'></a>*Euler number*

#####In mathematics, Euler's number, denoted as $ e $, is an irrational constant approximately equal to:

$$
e \approx 2.71828
$$

#####It is the base of the natural logarithm, and it arises naturally in many areas of mathematics, particularly in calculus.
\
##### Key Properties of $ e $:

1. **Limit Definition**:
   Euler's number can be defined as the limit of the following expression:

   $$
   e = \lim_{n \to \infty} \left( 1 + \frac{1}{n} \right)^n
   $$

2. **Exponential Function**:
   The exponential function $ e^x $ is one of the most important functions in mathematics and is defined as:

   $$
   e^x = \lim_{n \to \infty} \left( 1 + \frac{x}{n} \right)^n
   $$

   It has the unique property that its derivative is equal to the function itself:

   $$
   \frac{d}{dx} e^x = e^x
   $$

3. **Natural Logarithm**:
   The natural logarithm $ \ln(x) $ is the logarithm to the base $ e $, and it is the inverse of the exponential function:

   $$
   \ln(x) = \log_e(x)
   $$

###<a name='8'></a>*Limits*





#####In calculus, a **limit** is a fundamental concept that describes the behavior of a function as its argument approaches a particular point.

#####The limit of a function $ f(x) $ as $ x $ approaches a value $ c $ is written as:

$$
\lim_{x \to c} f(x) = L
$$

#####Where:
- $ x $ is the independent variable,
- $ c $ is the value that $ x $ approaches,
- $ L $ is the value that $ f(x) $ approaches as $ x $ approaches $ c $.

##### Types of Limits:
1. **Finite Limits:**
   $$ \lim_{x \to c} f(x) = L $$
   This means that the function approaches a finite value $ L $ as $ x $ approaches $ c $.

2. **Infinite Limits:**
   $$ \lim_{x \to c} f(x) = \infty $$
   This occurs when the function grows without bound as $ x $ approaches $ c $.

3. **Limits at Infinity:**
   $$ \lim_{x \to \infty} f(x) = L $$
   This describes the behavior of a function as $ x $ grows larger and larger.

##### Important Limit Laws:
- **Sum Law:**
   $$ \lim_{x \to c} (f(x) + g(x)) = \lim_{x \to c} f(x) + \lim_{x \to c} g(x) $$

- **Product Law:**
   $$ \lim_{x \to c} (f(x) \cdot g(x)) = \lim_{x \to c} f(x) \cdot \lim_{x \to c} g(x) $$

- **Quotient Law:**
   $$ \lim_{x \to c} \left( \frac{f(x)}{g(x)} \right) = \frac{\lim_{x \to c} f(x)}{\lim_{x \to c} g(x)} $$

- **Power Law:**
   $$ \lim_{x \to c} f(x)^n = \left( \lim_{x \to c} f(x) \right)^n $$

###<a name='9'></a>*Derivatives*

##### In calculus, a **derivative** is a fundamental concept that describes the rate of change of a function at a particular point.

##### The derivative of a function $ f(x) $ at a point $ x = c $ is written as:

$$
f'(c) = \lim_{h \to 0} \frac{f(c+h) - f(c)}{h}
$$

##### Where:
- $ f'(c) $ is the derivative of $ f(x) $ at $ x = c $,
- $ h $ is a small change in $ x $,
- $ f(c+h) - f(c) $ is the change in the function's value,
- The ratio $ \frac{f(c+h) - f(c)}{h} $ represents the slope of the secant line between two points on the curve, and as $ h $ approaches 0, this gives the slope of the tangent line at $ x = c $.

##### Types of Derivatives:
1. **First Derivative:**
   The first derivative of a function $ f(x) $ represents the rate of change or slope of the function. It is denoted as $ f'(x) $ or $ \frac{d}{dx} f(x) $.
   
   - Geometrically, it gives the slope of the tangent line to the curve at a point.
   
2. **Second Derivative:**
   The second derivative of a function $ f(x) $ is the derivative of the first derivative, written as $ f''(x) $ or $ \frac{d^2}{dx^2} f(x) $.
   
   - It measures the concavity of the function, or how the slope of the function is changing.

##### Common Rules for Derivatives:
- **Constant Rule:**
   If $ f(x) = c $ (where $ c $ is a constant), then:
   $$
   f'(x) = 0
   $$

- **Power Rule:**
   If $ f(x) = x^n $ (where $ n $ is a constant), then:
   $$
   f'(x) = n x^{n-1}
   $$

- **Sum Rule:**
   If $ f(x) = g(x) + h(x) $, then:
   $$
   f'(x) = g'(x) + h'(x)
   $$

- **Product Rule:**
   If $ f(x) = g(x) \cdot h(x) $, then:
   $$
   f'(x) = g'(x) \cdot h(x) + g(x) \cdot h'(x)
   $$

- **Quotient Rule:**
   If $ f(x) = \frac{g(x)}{h(x)} $, then:
   $$
   f'(x) = \frac{g'(x) \cdot h(x) - g(x) \cdot h'(x)}{h(x)^2}
   $$

- **Chain Rule:**
   If $ f(x) = g(h(x)) $, then:
   $$
   f'(x) = g'(h(x)) \cdot h'(x)
   $$

##### Calculate the derivative of the function $f(x) = 2x^2$ at $x = 3$ using the numerical approximation method:

#####Given the function:

$$
f(x) = 2x^2
$$

#####and using the derivative approximation formula:

$$
\text{Derivative} = \frac{f(x + h) - f(x)}{h}
$$

#####where $h$ is a small step size.

#####For this example, use a step size of $h = 0.1$.

#####Do not use any built-in libraries for differentiation

```python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / step_size
    return m

def my_function(x):
    return 2*x**2

slope_in_point_3 = derivative_x(my_function, 3, .01)

print(slope_in_point_3)
```

###<a name='10'></a>*Integrals*


#####In mathematics, the integral is one of the fundamental concepts of mathematical analysis, closely related to the concept of the derivative. Integrals are used to calculate many things, such as:

- Area under a curve,
- Motion transformations (in physics).

##### Types of Integrals:

1. **Indefinite Integral**  
   Represents a family of antiderivatives:
   $$
   \int f(x) \, dx
   $$
   The result includes a constant of integration $ C $.

2. **Definite Integral**  
   Represents the area under the curve of a function over an interval $[a, b]$:

   $$
   \int_a^b f(x) \, dx
   $$
   The result is a number.


#####Calculate the area under the curve $ f(x) = x^2 $ over the interval $[1, 3]$:
$$
\int_1^3 x^2 \, dx = \left[ \frac{x^3}{3} \right]_1^3 = \frac{26}{3}
$$
